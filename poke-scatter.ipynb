```python
def explode_rgb_scatter(img):
    fig = plt.figure(figsize=(14*0.75, 18*0.75))
    
    def rgb_scatter(fig, n, ijk):
        i, j, k = ijk
        ax = fig.add_subplot(3, 2, n, projection='3d', facecolor='white')
        colors = np.hstack([
            img[:, :, i].flatten().reshape(-1, 1),
            img[:, :, j].flatten().reshape(-1, 1),
            img[:, :, k].flatten().reshape(-1, 1)])
        ax.scatter(
            img[:, :, i].flatten(),
            img[:, :, j].flatten(),
            img[:, :, k].flatten(),
            c=img.reshape(-1, 4))
        rgb = ['R', 'G', 'B']
        l0, l1, l2 = rgb[i], rgb[j], rgb[k]
        ax.set_xlabel(l0)
        ax.set_ylabel(l1)
        ax.set_zlabel(l2)
        
    rgb_scatter(fig, 1, (0, 1, 2))
    rgb_scatter(fig, 2, (1, 2, 0))
    rgb_scatter(fig, 3, (2, 0, 1))
    rgb_scatter(fig, 4, (0, 2, 1))
    rgb_scatter(fig, 5, (2, 1, 0))
    rgb_scatter(fig, 6, (1, 0, 2))
    return fig
```

```python
fig = explode_rgb_scatter(bulbasaur)
```
